# Cleaning the Data

The Geography column has 2 fields seperated by comma so I have Split Geograghy into two columns - County and State 

```python
#dt.drop( columns = [ 'April 1, 2010 - Census' , 'April 1, 2010 - Estimates Base' ] , inplace = True) 
```

As Geography column has 2 fields,we are Splitting Geograghy into two columns - County and State 

```python
dt[[ 'County' , 'State' ]] = dt[ 'Geography' ].str.split( ', ' ,  expand=True )      # n=1 : for only 1 seperation 
```

```python
dt[ 'County' ] = dt[ 'County' ].str.rstrip( 'County' ) # Removing the word 'County' from rows in County  
```

```python
dt[ 'County' ] = dt[ 'County' ].str.rstrip( ' ' ) # Removing the space at the end from rows in County  
```

```python
dt.drop( columns = ["Geography"] , inplace = True)      # Dropping Geography Column

#To edit the original DataFrame, the “inplace” parameter can be set to True, and there is no returned value.
```

# Tidy Data
Merging the years into one column 'Year' and the value into column 'Population'

```python
dt.columns=dt.columns.str.lstrip( 'Population Estimate (as of July 1) -' )   #Before melting the data, Changing the column name to represent the year 
```

```python
dt=pd.melt(dt , [ 'Id' , 'Id2' , 'County' , 'State' , 'April 1, 2010 - Census' , 'April 1, 2010 - Estimates Base' ] , var_name='Year' , value_name='Population Estimate (as of July 1)' ) 
```

```python
dt.dtypes
```

```python
dt[ 'Year' ] = dt[ 'Year' ].astype(int)
```

# Visualisation

Plotting population in Scatter plot in Baltimore County for each year to see change in population in the county

```python
plt.scatter( x= dt[dt[ 'County' ] == 'Baltimore' ][ 'Year' ] , y = dt[dt[ 'County' ] == 'Baltimore' ][ 'Population Estimate (as of July 1)' ] )

#_=plt.ylim((0,max(dt[dt['County']=='Baltimore']['Population Estimate (as of July 1)'])))

plt.suptitle( 'Population Estimate in Baltimore County from year 2010 to 2018' )

plt.ylabel( 'Population Estimate' )

plt.xlabel( 'Year' )

plt.grid( True )
#_=plt.ylim((5000 + min(dt[dt['County']=='Baltimore']['Population Estimate (as of July 1)']),500 +max(dt[dt['County']=='Baltimore']['Population Estimate (as of July 1)'])))
```

Selecting Maryland State Population data for year 2010 and 2018 in m10 and m18 respectively

```python
m10 = dt[dt[ 'Year' ] == 2010]
m10 = m10[m10[ 'State' ] == 'Maryland' ]   # p10 is the population estimate in year 2010 in Maryland

m18 = dt[dt[ 'Year' ] == 2018]
m18 = m18[m18[ 'State' ] == 'Maryland']  # p18 is the population estimate in year 2010 in Maryland

m10=m10.sort_values(by=['Population Estimate (as of July 1)']) #Sorting data according to population 
m18=m18.sort_values(by=['Population Estimate (as of July 1)'])
```

Plotting Bar graph for population estimate of Maryland Counties in 2018

```python
plt.figure( figsize = (30,30) )

plt.bar( m18[ 'County' ] , m18[ 'Population Estimate (as of July 1)' ] )

_=plt.xlabel( 'Counties of Maryland ' )
_=plt.ylabel( 'Population Estimate ' )
_=plt.title( 'Population Estimate in Counties of Maryland in year 2018' )
```

Plotting Bar graph for comparison of population estimate between the year 2010 and 2018 of Maryland Counties

```python
ml10 = m10[ 'Population Estimate (as of July 1)' ].tolist()
ml18 = m18[ 'Population Estimate (as of July 1)' ].tolist()

newdf = pd.DataFrame( { '2010' : ml10 , '2018': ml18 } , index = m10['County'] )
ax = newdf.plot.bar()

ax.set_title( 'Population Estimate in Maryland State for Year 2010 and 2018' )
ax.set_ylabel( 'Population Estimate' )
```

Plotting Pie Chart of population distribution between Maryland Counties in 2010

```python
plt.figure( figsize = (15,20) )
plt.pie( m10 [ 'Population Estimate (as of July 1)' ] , labels = m10[ 'County' ] , shadow=True , autopct='%2.1f%%' )
plt.title('Population Percentage of Maryland Counties in 2010')
plt.show()
```

Plotting Pie Chart of population distribution between Maryland Counties in 2018

```python
plt.figure( figsize = (15,20))
plt.pie(m18['Population Estimate (as of July 1)'] , labels = m18['County'] , shadow=True , autopct='%2.1f%%' )
plt.title('Population Percentage of Maryland Counties in 2018')
plt.show()
```
